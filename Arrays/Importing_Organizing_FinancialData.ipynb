```python
# Connect to the world wide web using pandas_datareader 
from pandas_datareader import data as wb
```

```python
# Consider web sources as string values for us to specifically look at certain databases
# For this exercise, we are going to pull Proctor & Gamble's Stock Price Information dating back to 1995
# Use DataReader to read Yahoo Web Package and specifically looking at Proctor & Gamble's Stock (PG)
PG = wb.DataReader('PG', data_source='yahoo', start='1995-1-1')
```

```python
# Do this for five different companies/ticker symbols
tickers = ['PG','CHK','AA','T','NLY']
new_data = pd.DataFrame()
for t in tickers:
    new_data[t] = wb.DataReader(t, data_source = "yahoo", start = '1995-1-1')['Adj Close']
```
